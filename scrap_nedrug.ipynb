# Search nedrug.mfds.go.kr based on ATC Code and download excel



```
# import packages
import os
import json
import requests
import shutil
import pathlib as pl
import pandas as pd
```

```
# define save directory and query url
save_path="./download/"
nedrug_url="https://nedrug.mfds.go.kr/searchDrug/getExcel?sort=&sortOrder=&searchYn=&page=1&searchDivision=detail&itemName=&entpName=&ingrName1=&ingrName2=&ingrName3=&itemSeq=&stdrCodeName=&atcCodeName={}&indutyClassCode=&sClassNo=&narcoticKindCode=&cancelCode=&etcOtcCode=&makeMaterialGb=&searchConEe=AND&eeDocData=&searchConUd=AND&udDocData=&searchConNb=AND&nbDocData=&startPermitDate=&endPermitDate=&excelSearchCnt={}"

# query nedrug based on atc code and save result in .xlsx format
def query_api(atc_code):

	url = nedrug_url.format(atc_code,10000)

	if not os.path.exists(save_path):
		pl.Path(save_path).mkdir(parents=True, exist_ok=True)

	with open(save_path + atc_code + ".xlsx", 'wb') as handle:
	    response = requests.get(url, stream=True)

	    print("Now getting {}".format(atc_code))

	    if not response.ok:
	        print (response)

	    shutil.copyfileobj(response.raw, handle)
```

### Write down the ATC code to search and run the cell

```
atc_code = "R05DB27"  # this is the ATC Code to search
query_api(atc_code) 
df = pd.read_excel(save_path + atc_code + ".xlsx", header=0) # load saved excel file
df.head()  # show first 5 rows 
```
